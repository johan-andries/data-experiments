```python
labels = []
vectors = []
for input_line in open("work/location_vectors.csv"):
    input_list = input_line.strip().split(",")
    labels.append(input_list[0])
    vectors.append(list(map(lambda i: float(i), input_list[1:])))
```

```python
import numpy as np
from sklearn.manifold import TSNE
```

```python
X = np.array(vectors)
```

```python
model = TSNE(n_components=2, init="pca", method="exact") # , init="pca", 
np.set_printoptions(suppress=True)
projected = model.fit_transform(X) 
```

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (18,15)
```

```python
fig, ax = plt.subplots()
ax.scatter(projected.T[0], projected.T[1], s=0.1)

for i, txt in enumerate(labels):
    if txt in te_tonen:
        ax.annotate(txt, (projected.T[0][i],projected.T[1][i]))
```
